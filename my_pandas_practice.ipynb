```python
import pandas as pd
import numpy as np
df = pd.DataFrame([[4,9]]*3,columns=["A","B"])
print(df)

print("-----------------------")
print("dfapply")
print(df.apply(np.sqrt))

print("-----------------------")
print("column  sum")
print(df.apply(np.sum,axis=0))

print("-----------------------")
print("row  sum")
print(df.apply(np.sum,axis=1))

print("-----------------------")
print("row lambdaSeries")
print(df.apply(lambda x:[1,2],axis=1))

print("-----------------------")
print("row lambdaexpandlistdataframecolumn")
print(df.apply(lambda x:[1,2],axis=1,result_type="expand"))

print("-----------------------")
print("row lambdaresult_type=\"expand\"")
print("column nameseriesindex")
print(df.apply(lambda x:pd.Series([1,2],index=["foo","bar"]),axis=1))

print("-----------------------")
print("row lambdaresult_type=\"broadcast\"")
print("XXXXX")
print(df.apply(lambda x:[1,2],axis=1,result_type="broadcast"))
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,2*math.pi,100)
y=np.sin(x)

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("sinx")
plt.title("Sinx Function")
plt.show()
```
